## <u>Case Study</u>: Font Letter Images Dataset Analysis

In this analysis, we will explore a random sample of another popular machine learning benchmark image dataset known as the "notMNIST" dataset. This dataset is comprised of 500 28-by-28 pixel images of 10 letters, "A"-"J", each written in one of 50 different types of fonts. Each of the 784 image pixels is represented by a numerical gray scale value which can range from 0 (black) to 255 (white). Each object has two associated pre-assigned class labels: one which refers to what letter it is an image of, and one which refers to what font type it is.

The performance of many machine learning algorithms, especially when applied to image data, can be heavily influenced by the nature and efficacy of the data pre-processing steps that were taken. A course in computer vision would be able to more extensively discuss these specific image pre-processing steps to get better performance from our machine learning algorithms when applied to image data. For the purpose of this unsupervised learning class, this notMNIST dataset is a good dataset to start with, as these images do not require a lot of data pre-processing in order to get insightful results from our machine learning algorithms.

### 1. Data Preprocessing

This dataset does not have any missing values.

* Read the notMNIST_sample500.csv into a dataframe.
* Create another dataframe that contains just the scaled numerical pixel attributes. You should scale these pixel attributes by dividing each of the values by 255. **This is the dataset we will cluster**.
* Create another dataframe that contains the scaled numerical variables, the letter class labels, and the font class labels.

````python
df=pd.read_csv('notMNIST_sample500.csv')
df
````

````python
X=df.copy().drop(['letter', 'font'], axis=1)
X.head()

X = X/(255)
X
````

````python
df_stand=X.copy()
df_stand['letter']=df['letter']
df_stand['font']=df['font']
df_stand.head()
````

### 2. Basic Descriptive Analytics

#### 2.1. Number of Each Letter

Display the number of images of each type of letter.

````python
y=df['letter'].unique()
df['letter'].value_counts()
````

#### 2.2. Letter Images

Display the first ten images of each type of letter in the dataset.

````python
for digit in range(0,10):
    
    #Find just the rows in X that are an image of this particular digit
    df_digit_subset=X[df_stand['letter']==y[digit]]
    
    #Set up the plot
    fig, ax = plt.subplots(1, 10, figsize=(4, 2))
    plt.suptitle('Images of Digit: '+ str(y[digit]))
    
    #Visualize the first three images that are of this digit
    for i in range(0,10):
        #First, reshape the flattened 784 pixel array into an 28-by-28 numpy array and unscale the data (ie. multiply by 255)
        object_image = np.array(df_digit_subset.iloc[i]).reshape((28, 28))*255
        
        #Then show the image
        ax[i].imshow(object_image, cmap='gray')
        ax[i].grid(False)
        ax[i].set_xticks([])
        ax[i].set_yticks([])
    plt.show()
````

#### 2.3. Average Letter Image

Calculate the average of the images corresponding to each letter type. Then display the resulting averaged image for each of the 10 letter types.

````python
df_digit_averages=df.groupby(['letter']).mean()
df_digit_averages
````

````python
#Set up the plot
fig, ax = plt.subplots(1, 10, figsize=(10, 2))
plt.suptitle('Averaged Image of Each Digit')

for digit in range(0,10):
    #First, reshape the flattened 784 pixel array into an 28-by-28 numpy array and unscale the data (ie. multiply by 255)
    object_image = np.array(df_digit_averages.loc[y[digit]]).reshape((28, 28))*255

    #Then show the image
    ax[digit].imshow(object_image, cmap='gray')
    ax[digit].grid(False)
    ax[digit].set_xticks([])
    ax[digit].set_yticks([])
    ax[digit].set_title('Digit %s'%str(y[digit]))
plt.show()
````

### 3. Clusterability

#### 3.1. t-SNE Plots
Using 6 different perplexity values and at least two random states for each perplexity value, map this 784 dimensional dataset onto a two-dimensional dataset with the t-SNE algorithm. Show your projected coordinates in a scatterplot for each combination of random states and perplexity value. Also, color code your points by the letter labels.

````python
for perp in [5,10, 20, 30, 40, 50]:
    for rs in [103,104]:
        tsne = TSNE(n_components=2, perplexity=perp, random_state=rs)
        data_tsne = tsne.fit_transform(X)
        df_tsne = pd.DataFrame(data_tsne, columns=['x_projected', 'y_projected'])
        df_combo = pd.concat([df, df_tsne], axis=1)
        sns.scatterplot(x='x_projected',y='y_projected', hue='letter', palette=sns.color_palette("husl", 10), data=df_combo)
        plt.title('t-SNE Plot with Perplexity Value %s and Random State %s' %(perp, rs))
        plt.legend(bbox_to_anchor=(1,1))
        plt.show()
    print('--------------------------------------------')
````

#### 5.5. Parameter Selection

One of the goals of this cluster analysis is to return a fuzzy clustering result that:
* is able to identify the underlying clusters in the dataset,
* is able to reflect the uncertain nature of which clusters "straddle nodes" should belong to, and
* is able to reflect the more certain nature of which cluster highly cohesive objects belong to.

Which parameter value (p=1.01, 1.1, 1.7, or 2) creates the fuzzy clustering that best meets these goals? Explain. 

The clustering with p=1.01 makes very hard assignments. From the corresponding histogram, we see that almost all objects have their highest cluster membership weight very close to 1. So, a very hard assignment like this will not easily indicate the presence of very many straddle nodes or reflect any underlying "fuzziness" of the clusters that may exist. 

Then, when p=1.7 and p=2, the clustering result lumps all clusters into one large, fuzzy, indistiguishible cluster. Tis clustering does not allow us to easily identify which cluster most of the objects in the dataset should belong to. So this is not the most useful set of parameters.

However, with p=1.1, some nodes that may be on the "outskirts" of their respective clusters have fuzzier cluster membership weights. From the histogram, we can see that many objects have their highest cluster membership weight very close to 1. But there are also quite a few that have much lower highest cluster membership weights, indicating that they may be straddle nodes to varying degrees. Therefore, it meets the goals to show which cluster "straddle nodes" should belong to.

````python
#Number of clusters
c=10

#List of cluster membership score matrices for each p
W_list = []

#List of normalized dunning partitional coefficient scores for each p
ndpc_list = []

#List of cluster centroids for each p
centroids_list=[]

for p in (1.01, 1.1, 1.7, 2):
    print('------------------------------------------')
    print('Analysis for c=%s and p=%s'%(c,p))

    
    #Cluster the data using these parameters
    centroids, W, _, _, obj_history, _, ndpc = cmeans(X.T, c=10, m=p, error=1e-6, maxiter=20000, seed=100)

    #Normalized Dunning Partitional Coefficient
    print('Normalized Dunning Partitional Coefficient: ', ndpc)
    print('Final Objective Function Value: ' , obj_history[len(obj_history)-1])
    
    #Update the lists
    W_list.append(W)
    ndpc_list.append(ndpc)
    centroids_list.append(centroids)
    
    #Color code t-SNE Plots
    fig, ax = plt.subplots(2, 5, figsize=(20, 10))
    plt.suptitle('Fuzzy Clustering: %s clusters and p=%s'%(c,p), fontsize=14)
    for i in range(0,10):
        
        #Create a column with cluster i membership scores
        df_combo['Cluster %s Membership Score'%str(i)]=W[i,:]

        #Plot the points, color coded by the membership score.
        points = plt.scatter(df_combo["x_projected"], df_combo["y_projected"],
                     c=df_combo['Cluster %s Membership Score'%str(i)],
                             s=30, cmap="BuGn",
                            vmin=0,
                            vmax=1)
        ax[int(np.floor(i/5)),int(i%5)].scatter(df_combo["x_projected"], df_combo["y_projected"],
                     c=df_combo['Cluster %s Membership Score'%str(i)],
                                                s=30, cmap="BuGn",
                                               vmin=0,
                                               vmax=1)

        fig.colorbar(points, ax=ax[int(np.floor(i/5)),int(i%5)])
        ax[int(np.floor(i/5)),int(i%5)].set_title('Cluster %s Membership Scores'%(str(i)))
    plt.show()
    
    
    #Let's visualize and describe the cluster centroids for this clustering
    #Set up the plot
    fig, ax = plt.subplots(1, 10, figsize=(12, 2))
    plt.suptitle('Images of the Cluster Centroids')

    for j in range(0,10):
        #Turn the centroids into a dataframe
        df_centroids=pd.DataFrame(centroids, columns=X.columns)
        #First, reshape the flattened 784 pixel array into an 28-by-28 numpy array and unscale the data (ie. multiply by 255)
        object_image = np.array(df_centroids.iloc[j]).reshape((28, 28))*255

        #Then show the image
        ax[j].imshow(object_image, cmap='gray')
        ax[j].grid(False)
        ax[j].set_xticks([])
        ax[j].set_yticks([])
        ax[j].set_title('Centroid %s'%str(j))
    plt.show()
    
    #Let's get the highest membership score for the clustering for each object in the dataset
    cluster_columns=['Cluster %s Membership Score'%(str(i)) for i in range(0,10)]
    df_membership=pd.DataFrame(W.T, columns=cluster_columns)
    highest_membership_scores=df_membership.T.max()
    
    print('Highest Cluster Membership Scores (Sorted)')
    print(highest_membership_scores.sort_values())
    
    if highest_membership_scores.max()>highest_membership_scores.min()+.00001:
        plt.hist(highest_membership_scores, bins=20)
        plt.title('Highest Membership Score Value for an Object in the Clustering')
        plt.xlabel('Highest Membership Score Value for an Object')
        plt.ylabel('Frequency')
        plt.xlim([0,1])
        plt.show()
````
