## 3. Gaussian Mixture Models Parameter Selection

We would like to use Gaussian Mixture models to cluster this dataset. However, we are currently unsure as to how many clusters we should ask for.

### 3.1. Parsimonious Gaussian Mixture Model

We would first like to assess the number of clusters $k$ which will yield a parsimonious Gaussian mixture model for this dataset. 

#### 3.1.1. Reliability of AIC for Selecting a Parsimonious Model

For what values of the cluster number $k$ can we rely on the AIC score to produce useful insights regarding the parsimonious model? In general, can we rely on the AIC score at all with this dataset?



A rule of thumb indicates that it's best to use AIC when $\frac{n}{n_p}>40$.

Thus because $n=500$, then we should use AIC to make assessments regarding a parsimonious number of parameters only in the range of $n_p<12.5$.

Because in our Gaussian mixture models, the number of parameters that we need in our model is:
$n_p = k(1 + d+ d^2)$ (where d=784=number of attributes), then $k<0.00002$. Thus using AIC is not ideal for assessing whether any value of $k$ yields a parsimonious model.

````python
# Calculate Np
500/40
````

#### 3.1.2. AIC and BIC Score Plots

Create two plots.
1. One plot should plot the AIC score of a Gaussian mixture model clustering with k=1, k=2, ..., k=19 clusters respectively.
2. The other plot should plot the BIC score of a Gaussian mixture model clustering with k=1, k=2, ..., k=19 clusters respectively.

Use a random state of 101 for all of these clusterings.

````python
aics = []
bics = []

for k in range(1, 25):
    #Fit the GMM model
    gm = GaussianMixture(n_components=k, random_state=100)
    gm.fit(X)
    
    #Calculate the AIC and BIC
    aics.append(gm.aic(X))
    bics.append(gm.bic(X))


#Plot these values
fig, ax = plt.subplots(1, 2, figsize=(30, 6))

#AIC
ax[0].plot(range(1, 25), aics)
ax[0].set_xticks(range(1, 25))
ax[0].set_xlabel('k', fontsize=16)
ax[0].set_title('AIC', fontsize=16)

#BIC
ax[1].plot(range(1, 25), bics)
ax[1].set_xticks(range(1, 25))
ax[1].set_xlabel('k', fontsize=16)
ax[1].set_title('BIC', fontsize=16)

plt.show()
````

#### 3.1.3 Interpretation

Use the AIC score to select the number of clusters that will yield a parsimonious Gaussian mixture model. Use the BIC score to select the number of clusters that will yield a parsimonious Gaussian mixture model.

The AIC score is suggesting that a Gaussian mixture model clustering with k=2 clusters will yield a parsimonious model. The BIC score is suggesting that a Gaussian mixture model clustering with k=1 clusters will yield a parsimonious model.


#### 3.1.4. Usefulness

Given that our goal is to determine the underlying clusters of this dataset, particularly those that the t-SNE plot is suggesting might exist, do we think that the cluster number $k$ suggested by the AIC or BIC score are useful?

The cluster number of k=2 suggested by the AIC score is more useful because the t-SNE plot does show 2 distinct clusters that have overlap. THe BIC score is not useful because grouping all the objects into 1 cluster is counterintuitive.

## 4. Gaussian Mixture Model with k=10 clusters

### 4.1. Cluster the Data
Cluster the dataset into k=10 clusters using Gaussian mixture models. Use a random state of 101.

````python
#Instantiate the GMM object gm (with your desired algorithm parameters)
gm = GaussianMixture(n_components=10, random_state=101)

#Run the GMM-EM algorithm on your data
gm.fit(X)

#Get the cluster membership scores (ie. posterior probabilities, P(x in Cluster_k| x)))
W=gm.predict_proba(X)
print(np.shape(W))
W
````

````python
cluster_columns=['Cluster %s Membership Score'%(str(i)) for i in range(0,10)]
df_membership=pd.DataFrame(W, columns=cluster_columns)
df_membership.round(4)
````

### 4.2. t-SNE Plots

Then, plot k=10 t-SNE plots. In the kth plot, you should color-code the objects by their kth cluster membership weight (ie. their kth cluster posterior value).

````python
#Color code t-SNE Plots
fig, ax = plt.subplots(2, 5, figsize=(20, 10))
plt.suptitle('Fuzzy Clustering: %s clusters'%(10), fontsize=14)
for i in range(0,10):

    #Create a column with cluster i membership scores
    df_combo['Cluster %s Membership Score'%str(i)]=df_membership['Cluster %s Membership Score'%(str(i))]

    #Plot the points, color coded by the membership score.
    points = plt.scatter(df_combo["x_projected"], df_combo["y_projected"],
                 c=df_combo['Cluster %s Membership Score'%str(i)],
                         s=30, cmap="BuGn",
                        vmin=0,
                        vmax=1)
    ax[int(np.floor(i/5)),int(i%5)].scatter(df_combo["x_projected"], df_combo["y_projected"],
                 c=df_combo['Cluster %s Membership Score'%str(i)],
                                            s=30, cmap="BuGn",
                                           vmin=0,
                                           vmax=1)

    fig.colorbar(points, ax=ax[int(np.floor(i/5)),int(i%5)])
    ax[int(np.floor(i/5)),int(i%5)].set_title('Cluster %s Membership Scores'%(str(i)))
plt.show()
````

**Cluster 1**:
